```python
from hero import Hero
from ability import Ability
from armor import Armor
from weapon import Weapon
# Create Hero instances
my_hero = Hero("Goku", 2000)
my_opponent = Hero("Vegeta", 2500)

# Print Hero details
print(my_hero.name)
print(my_hero.current_health)

# Add Abilities to Heroes
punch = Ability("Punch", 150)
kick = Ability("Kick", 200)

my_hero.add_ability(punch)
my_opponent.add_ability(kick)

# Add Armor to Heroes
shield = Armor("Shield", 50)
helmet = Armor("Helmet", 30)

my_hero.add_armor(shield)
my_opponent.add_armor(helmet)

# Test ability attributes
print(f"Ability: {punch.name}, Max Damage: {punch.max_damage}")

# Test attack method
print(f"Attack damage from {my_hero.name}: {my_hero.attack()}")
print(f"Attack damage from {my_opponent.name}: {my_opponent.attack()}")

# Test defend method
print(f"Defense block from {my_hero.name}: {my_hero.defend()}")
print(f"Defense block from {my_opponent.name}: {my_opponent.defend()}")

# Start a battle and print the result
my_hero.battle(my_opponent)

# Test Hero after battle
print(f"Health of {my_hero.name} after battle: {my_hero.current_health}")
print(f"Health of {my_opponent.name} after battle: {my_opponent.current_health}")

# Additional Tests
# Test adding multiple abilities and armors
fireball = Ability("Fireball", 300)
lightning = Ability("Lightning", 250)

my_hero.add_ability(fireball)
my_hero.add_ability(lightning)

heavy_armor = Armor("Heavy Armor", 100)
gauntlet = Armor("Gauntlet", 75)

my_hero.add_armor(heavy_armor)
my_hero.add_armor(gauntlet)

# Test attack with multiple abilities
print(f"Total defense block from {my_hero.name} with multiple armors: {my_hero.defend()}")

# Add Weapons to Heroes
sword = Weapon("Sword", 300)
axe = Weapon("Axe", 250)

my_hero.add_weapon(sword)
my_opponent.add_weapon(axe)

# Test attack with weapons
print(f"Attack damage from {my_hero.name}: {my_hero.attack()}")
print(f"Attack damage from {my_opponent.name}: {my_opponent.attack()}")
 # Display initial hero details
print(f"{my_hero.name}'s Initial Health: {my_hero.current_health}")
print(f"{my_opponent.name}'s Initial Health: {my_opponent.current_health}")

# Start battle
print("\n-Battle Begins-\n")
my_hero.battle(my_opponent)

# Display results after battle
print("\n--- Battle Results ---\n")
print(f"{my_hero.name}'s Health After Battle: {my_hero.current_health}")
print(f"{my_opponent.name}'s Health After Battle: {my_opponent.current_health}")

# Test additional scenarios
print("\nAdditional Tests \n")

# Test adding multiple abilities and armors to my_hero
print("\nAdding Multiple Abilities and Armors to Goku:\n")
print("Goku's Abilities:")
for ability in my_hero.abilities:
    print(f"{ability.name}: Max Damage - {ability.max_damage}")

print("\nGoku's Armors:")
for armor in my_hero.armors:
    print(f"{armor.name}: Block - {armor.block}")

# Test attack with multiple abilities
print(f"\nTotal Defense Block from {my_hero.name} with Multiple Armors: {my_hero.defend()}")

# Test attack with weapons
print(f"\nAttack Damage from {my_hero.name} with Weapons: {my_hero.attack()}")
print(f"Attack Damage from {my_opponent.name} with Weapons: {my_opponent.attack()}")



    
```
